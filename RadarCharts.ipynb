```python
import pandas as pd
import numpy as np


df=pd.read_html('https://fbref.com/en/players/9c36ed83/Nicolas-Jackson', attrs={'id':"stats_shooting_dom_lg"})[0]
```

```python
df=df.dropna()
```

```python
df_filtered = df[df['Unnamed: 0_level_0_Season'] == '2024-2025']
```

```python
#create garnacho dataframe

G_df=pd.read_html('https://fbref.com/en/players/7aa8adfe/Alejandro-Garnacho',attrs={'id':"stats_shooting_dom_lg"})[0]
```

```python
G_df.head()
```

```python
G_df.columns=['_'.join(col).strip() for col in G_df.columns.values]
```

```python
G_df.columns
```

```python
g_filtered=G_df[G_df['Unnamed: 0_level_0_Season']=='2023-2024']
```

```python
g_filtered
```

```python
#combine the two dataframes

combined_df=pd.concat([g_filtered,df_filtered],ignore_index=True)
```

```python
combined_df
```

```python
combined_df.columns
```

```python
new_df=combined_df.drop(columns=['Unnamed: 0_level_0_Season','Unnamed: 1_level_0_Age','Unnamed: 2_level_0_Squad','Unnamed: 3_level_0_Country','Unnamed: 4_level_0_Comp',
                         'Unnamed: 5_level_0_LgRank','Expected_G-xG','Expected_np:G-xG','Unnamed: 24_level_0_Matches','Standard_FK','Standard_PK','Standard_PKatt'])

                         
```

```python
new_df
```

```python
parameters=new_df.columns
```

```python
parameters
```

```python
ranges=[]
a_values=[]
b_values=[]

for param in parameters:
    new_df[param]=pd.to_numeric(new_df[param], errors='coerce')

#find minimum and maximum values for the parameters
    min_val=new_df[param].min()
    max_val=new_df[param].max()


#calculate range with 25% padding

    a=min_val-(min_val*0.25)
    b=max_val +(max_val*0.25)

    ranges.append((a,b))

jackson_data=new_df[new_df['Unnamed: 6_level_0_90s']==28.5].values.tolist()
garnacho_data=new_df[new_df['Unnamed: 6_level_0_90s']==19.3].values.tolist()


a_values=jackson_data[0] if jackson_data else None
b_values =garnacho_data[0] if garnacho_data else None


```

```python
from mplsoccer import Radar, FontManager, grid
import matplotlib.pyplot as plt


min_range, max_range=zip(*ranges)
min_range=list(min_range)
max_range=list(max_range)





radar = Radar(parameters, min_range, max_range, lower_is_better=[],
              
              # whether to round any of the labels to integers instead of decimal places
              round_int=[False]*len(parameters),
              num_rings=4,  # the number of concentric circles (excluding center circle)
              # if the ring_width is more than the center_circle_radius then
              # the center circle radius will be wider than the width of the concentric circles
              ring_width=1, center_circle_radius=1)



URL1 = ('https://raw.githubusercontent.com/googlefonts/SourceSerifProGFVersion/main/fonts/'
        'SourceSerifPro-Regular.ttf')
serif_regular = FontManager(URL1)
URL2 = ('https://raw.githubusercontent.com/googlefonts/SourceSerifProGFVersion/main/fonts/'
        'SourceSerifPro-ExtraLight.ttf')
serif_extra_light = FontManager(URL2)
URL3 = ('https://raw.githubusercontent.com/google/fonts/main/ofl/rubikmonoone/'
        'RubikMonoOne-Regular.ttf')
rubik_regular = FontManager(URL3)
URL4 = 'https://raw.githubusercontent.com/googlefonts/roboto/main/src/hinted/Roboto-Thin.ttf'
robotto_thin = FontManager(URL4)
URL5 = ('https://raw.githubusercontent.com/google/fonts/main/apache/robotoslab/'
        'RobotoSlab%5Bwght%5D.ttf')
robotto_bold = FontManager(URL5)




# Create figure and axes
fig, ax = radar.setup_axis()

# Plot player data
#radar.draw_circles(ax=ax)
player1_radar=radar.draw_radar(a_values, ax=ax, kwargs_radar={'color': 'blue', 'alpha': 0.3})
player2_radar=radar.draw_radar(b_values, ax=ax, kwargs_radar={'color': 'red', 'alpha': 0.5})

# Add labels
radar.draw_range_labels(ax=ax, fontsize=12)
radar.draw_param_labels(ax=ax, fontsize=14)


ax.legend([player1_radar[0], player2_radar[0]], ["Player 1", "Player 2"], loc='upper right', fontsize=12)

plt.savefig("radar_chart.png", dpi=300, bbox_inches='tight') 

# Show plot
plt.show()
```
